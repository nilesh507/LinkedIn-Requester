```python
df_2023 = pd.read_csv('./data/h1b_datahubexport-2023.csv')
df_2022 = pd.read_csv('./data/h1b_datahubexport-2022.csv')
df_2021 = pd.read_csv('./data/h1b_datahubexport-2021.csv')
df_2020 = pd.read_csv('./data/h1b_datahubexport-2020.csv')
df_2019 = pd.read_csv('./data/h1b_datahubexport-2019.csv')
df = pd.concat([df_2023, df_2022, df_2021, df_2020, df_2019])
df
```

```python
subset = df[['Fiscal Year', 'Employer', 'Initial Approval', 'Initial Denial', 'State', 'City', 'ZIP']]
subset
```

```python
#approval and denial for the last 3 years
approval = subset[subset['Fiscal Year'].isin([2023,2022,2021])]['Initial Approval'].sum()
denial = subset[subset['Fiscal Year'].isin([2023,2022,2021])]['Initial Denial'].sum()
approval,denial
```

```python
# Top states with highest H-1B visa approvals in the last 3 years
subset.groupby('State')['Initial Approval'].sum().sort_values(ascending=False)
```

```python
# Top cities with highest H-1B visa approvals in the last 3 years
subset.groupby('City')['Initial Approval'].sum().sort_values(ascending=False)
```

```python
# top employers, as in the employers with the highest approvals in the last 3 years.
subset.groupby('Employer')['Initial Approval'].sum().sort_values(ascending=False).head(100)
```

```python
def scroll_to_bottom(driver):
    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    match=False
    while(match==False):
        lastCount = lenOfPage
        time.sleep(0.1)
        lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
        if lastCount==lenOfPage:
            match=True
```

## Extracting the list of people to contact

```python
data = BeautifulSoup(driver.page_source, 'html.parser')
pages = data.find(class_="artdeco-pagination__pages artdeco-pagination__pages--number")
df = pd.DataFrame(columns =['Name', 'Title','Location', 'Link'])

# for i in range( (len(pages.find_all('li'))) if pages else 1):
i=0
while True and i<5:
    scroll_to_bottom(driver)
    data = BeautifulSoup(driver.page_source, 'html.parser')
    peoples = data.find_all(class_="entity-result__content entity-result__divider pt3 pb3 t-12 t-black--light")
    list = []
    for people in peoples:
        detail = people.find(class_="entity-result__title-text")
        link = detail.a.get('href')
        name = detail.a.span.span.get_text()
        title = people.find(class_="entity-result__primary-subtitle t-14 t-black t-normal")
        if  title:
            title = title.get_text()
        else:
            title = "Not Mentioned"
        location = people.find(class_="entity-result__secondary-subtitle t-14 t-normal").get_text()
        list.append([name.strip(), title.strip(), location.strip(), link.strip()])
    temp = pd.DataFrame(list, columns =['Name', 'Title','Location', 'Link'])
    df = pd.concat([df,temp], ignore_index=True)
    if not pages :
        print('broken from page')
        break
    next = driver.find_element(By.XPATH,'//button[@aria-label="Next"]')
    if not next.is_enabled():
        print('broken from next')
        break
    i += 1
    next.click()
    time.sleep(1.5)
```
